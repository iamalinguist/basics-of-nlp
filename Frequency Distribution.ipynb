# To Tabulate the Occurances as per conditions

```python
# First we create universal tuple [consisting of (genre,word)]
```

```python
cfd = ConditionalFreqDist(
           (genre, word) #create tuple of X,Y
          for genre in brown.categories() #take genre nothing but categories from corpus
          for word in brown.words(categories=genre)) #take words as per categories from corpus


#Remember Above Tuples Contain All Data, So below You have to choose your Xreq, Yreq
#Don't give the same name as tuple to below X and Y
Xreq = ['news', 'religion'] #lets choose genre we are interested in
Yreq = ['can', 'could', 'may', 'might', 'must'] #lets choose words we are interested in
cfd.tabulate(conditions=Xreq, samples=Yreq)
```
